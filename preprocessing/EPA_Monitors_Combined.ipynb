```python
import glob
import pandas as pd
```

```python
file_list = glob.glob("../data/**/*combined*.csv", recursive=True)
```

```python
file_list
```

```python
import matplotlib.pyplot as plt

for i, f in enumerate(file_list):
    # do something with the combined file
    df = pd.read_csv(f, index_col='date')
    df = df.set_index(pd.DatetimeIndex(df.index))
    df = df.rename(columns={'sample_measurement':f.split('\\')[-1].split('_')[0]})

    if i == 0:
        
        ax = df.resample('M').mean().plot(kind='line',figsize=(12, 4), rot=45)
        ax.legend(bbox_to_anchor=(0.5, -0.20), loc='upper center',fancybox=True,shadow=True,ncol=5)
    else:
        df.resample('M').mean().plot(ax=ax, kind='line', rot=45)
        
        plt.title('Colorado EPA PM$_{2.5}$ Monitor Data from 2017 to 2023',fontsize=17)
        plt.xlabel('Time',fontsize=10)
        plt.ylabel('PM$_{2.5}$ ( $\mu$g / m$^3$)',fontsize=10)
        
        ax.legend(bbox_to_anchor=(0.5, -0.20), loc='upper center',fancybox=True,shadow=True,ncol=5)
        #plt.savefig('AllEPAMonitors20172022.png',bbox_inches='tight',dpi=200)
```
